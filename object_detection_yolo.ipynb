```python
from ultralytics import YOLO
import cv2
import os
import numpy as np

# Function to calculate Intersection over Union (IoU)
def iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    denom = float(boxAArea + boxBArea - interArea)
    return interArea / denom if denom > 0 else 0.0

# Image Sharpening Functions
def unsharp_mask(image, kernel_size=(5, 5), sigma=1.0, amount=1.0, threshold=0):
    blurred = cv2.GaussianBlur(image, kernel_size, sigma)
    sharpened = float(amount + 1) * image - float(amount) * blurred
    sharpened = np.maximum(sharpened, np.zeros(sharpened.shape))
    sharpened = np.minimum(sharpened, 255 * np.ones(sharpened.shape))
    sharpened = sharpened.round().astype(np.uint8)
    if threshold > 0:
        low_contrast_mask = np.absolute(image - blurred) < threshold
        np.copyto(sharpened, image, where=low_contrast_mask)
    return sharpened

def laplacian_sharpen(image):
    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    sharpened = cv2.convertScaleAbs(image + laplacian)
    return sharpened

def kernel_sharpen(image):
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    sharpened = cv2.filter2D(image, -1, kernel)
    return sharpened

# ------------------------------------------------------------
# Configuration
# ------------------------------------------------------------
model_path = r"C:\Users\joshu\Documents\EEE\AMNIS\Models\training\weights\best.pt"  # Path to YOLO model weights
video_path = r"C:\Users\joshu\Queen's University Belfast\Michael Loughran - 20241204_jp\20241204_104958_BR1.mp4"
save_dir = r"C:\Users\joshu\Documents\EEE\AMNIS\output_videos"  # Directory to save output video
os.makedirs(save_dir, exist_ok=True)

# Thresholds
conf_thresh = 0.1
nms_iou_thresh = 0.3       # Renamed for clarity
tracking_iou_thresh = 0.3  # Renamed for clarity
shelf_thresh = 100  # Threshold for grouping shelves by bottom edge
lost_time_limit = 300  # Number of frames to keep lost objects

# Load YOLO model
model = YOLO(model_path)

# Initialize video capture
cap = cv2.VideoCapture(video_path)
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Allow user to draw ROI
ret, first_frame = cap.read()
if not ret:
    print("Error: Could not read the first frame.")
    exit()
roi = cv2.selectROI("Select ROI", first_frame, fromCenter=False, showCrosshair=True)
cv2.destroyWindow("Select ROI")
x_min, y_min, w, h = roi
x_max, y_max = x_min + w, y_min + h

# Video writer
output_video_path = os.path.join(save_dir, "tracked_with_shelves_roi_fixed.mp4")
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out_writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Initialize tracker state
active_objects = {}  # {id: box}
lost_objects = {}    # {id: (box, frame_last_seen)}
next_label_num = 1
objects_that_left = set()
frame_count = 0
initial_object_count = None  # Will set after first frame with objects
color = (0, 255, 0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_count += 1

    # Apply Image Sharpening
    # Choose one of the sharpening methods below:
    # 1. Unsharp Mask
    sharpened_frame = unsharp_mask(frame, kernel_size=(5,5), sigma=1.0, amount=1.5, threshold=0)
    
    # 2. Laplacian Sharpening
    # sharpened_frame = laplacian_sharpen(frame)
    
    # 3. Kernel-Based Sharpening
    # sharpened_frame = kernel_sharpen(frame)
    
    # Run YOLO prediction on the sharpened frame
    results = model.predict(sharpened_frame, conf=conf_thresh, iou=nms_iou_thresh)
    detections = results[0].boxes

    # Extract bounding boxes within ROI
    boxes = []
    for det in detections:
        x1, y1, x2, y2 = map(int, det.xyxy[0])
        if (x_min <= x1 <= x_max and x_min <= x2 <= x_max and
            y_min <= y1 <= y_max and y_min <= y2 <= y_max):
            boxes.append((x1, y1, x2, y2))

    if boxes:
        # Group boxes into shelves by bottom edge
        boxes.sort(key=lambda b: b[3])
        shelves = []
        for box in boxes:
            x1, y1, x2, y2 = box
            placed = False
            for shelf in shelves:
                shelf_mean_bottom = sum(b[3] for b in shelf) / len(shelf)
                if abs(y2 - shelf_mean_bottom) <= shelf_thresh:
                    shelf.append(box)
                    placed = True
                    break
            if not placed:
                shelves.append([box])

        # Sort shelves and flatten the list
        for shelf in shelves:
            shelf.sort(key=lambda b: b[0])
        sorted_boxes = [b for shelf in shelves for b in shelf]

        # Object tracking logic
        new_objects = {}
        matched_ids = set()

        for new_box in sorted_boxes:
            best_iou_score = 0
            best_id = None

            # Match with active objects
            for lbl, a_box in active_objects.items():
                iou_val = iou(a_box, new_box)
                if iou_val > best_iou_score:
                    best_iou_score = iou_val
                    best_id = lbl

            # If no good match with active_objects, try lost_objects
            if best_iou_score < tracking_iou_thresh:
                for lbl, (lost_box, lost_frame) in list(lost_objects.items()):
                    iou_val = iou(lost_box, new_box)
                    if iou_val > best_iou_score:
                        best_iou_score = iou_val
                        best_id = lbl

                # If matched with a lost object, restore it
                if best_iou_score >= tracking_iou_thresh and best_id in lost_objects:
                    del lost_objects[best_id]

            # If still no match found, assign a new ID
            if best_iou_score < tracking_iou_thresh:
                best_id = next_label_num
                next_label_num += 1

            new_objects[best_id] = new_box
            matched_ids.add(best_id)

        # Determine which objects disappeared
        current_ids = set(new_objects.keys())
        old_ids = set(active_objects.keys())
        disappeared = old_ids - current_ids

        for d_id in disappeared:
            if d_id not in lost_objects:
                lost_objects[d_id] = (active_objects[d_id], frame_count)
                objects_that_left.add(d_id)

        # Clean lost_objects
        lost_objects = {lid: (lbox, lframe) for lid, (lbox, lframe) in lost_objects.items() 
                        if frame_count - lframe <= lost_time_limit}

        active_objects = new_objects

        # If an object that was previously considered left reappears, remove it from objects_that_left
        for oid in matched_ids:
            if oid in objects_that_left:
                objects_that_left.remove(oid)

        # If we have not yet set the initial object count, do it now
        if initial_object_count is None:
            initial_object_count = len(active_objects)

        # Draw active objects
        for lbl, (x1, y1, x2, y2) in active_objects.items():
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"{lbl}", (x1 + 5, y1 + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        # Draw shelves
        for idx, shelf in enumerate(shelves):
            y_bottom = min(b[3] for b in shelf)
            cv2.line(frame, (0, y_bottom), (width, y_bottom), (0, 0, 255), 2)
            cv2.putText(frame, f"Shelf {idx + 1}", (10, y_bottom - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    else:
        # If no boxes this frame, all current active objects might be lost
        if active_objects:
            for lbl, box in active_objects.items():
                if lbl not in lost_objects:
                    lost_objects[lbl] = (box, frame_count)
                    objects_that_left.add(lbl)
        active_objects = {}

        # Clean lost_objects
        lost_objects = {lid: (lbox, lframe) for lid, (lbox, lframe) in lost_objects.items()
                        if frame_count - lframe <= lost_time_limit}

    # Draw ROI rectangle
    cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)

    # Display objects that have left
    if objects_that_left:
        text = f"Objects that have left: {', '.join(map(str, objects_that_left))}"
        cv2.putText(frame, text, (100, 90), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

    # Display current total objects in scene (if initial_object_count is known)
    if initial_object_count is not None:
        current_count = initial_object_count - len(objects_that_left)
        info_text = f"Total objects currently in scene: {current_count}/{initial_object_count}"
        cv2.putText(frame, info_text, (100, 130), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

    # Write frame
    out_writer.write(frame)

cap.release()
out_writer.release()
print(f"Processed video saved to {output_video_path}")
```
